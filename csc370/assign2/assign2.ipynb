1. $A \rightarrow B$, $BC \rightarrow D$, $A \rightarrow C$

A relation is in BCNF if for each non-trivial FD($X\leftarrow Y$) that holds in the relation, it is the case that the LHS ($X$) is a superkey.  

$\{ A^+ \}: A \Rightarrow AB \Rightarrow ABC \Rightarrow ABCD$  which means the closure of A is a super key, so $A \rightarrow B$ and $A \rightarrow C$ are not in violation of BCNF.  
$\{ BC^+ \}: BC \Rightarrow BCD$  which means the closure of $BC$ is not a super key, so $BC \rightarrow D$ is in violation of BCNF.  

Thus $R(ABCD)$ would not be in BCNF with FDs $A \rightarrow B$, $BC \rightarrow D$, $A \rightarrow C$.  

So now we will decompose R into relations that are in BCNF.  

As $BC \rightarrow D$ is an FD in violation, we will sort it out by settings $R$ into $R1$ and $R2$ as follows...  
Using the FD in violation we get $R1(BCD)$.  
$R1(BCD)$ $\Rightarrow$ $(R - \{D\}) = R2$, so $R2(ABC)$.  

We have exhausted of the original FDs in violation from R. We will check again to see if any remaining FDs are in violation just to be sure.  

$\{ A^+ \}: A \Rightarrow AB \Rightarrow ABC$  which means the closure of A is a super key for $R2$ (all relations it can act on), so $A \rightarrow B$ and $A \rightarrow C$ are not in violation of BCNF.   
$\{ BC^+ \}: BC \Rightarrow BCD$  which means the closure $BC$ is a super key for $R1$ (all relations it can act on), so $BC \rightarrow D$ is not in violation of BCNF.  

**Resulting Relations**  
$R1(BCD)$  
$R2(ABC)$  

**Loseless Join?**

1. What are all the keys for $Stocks$?

$\{ C^+ \}: C \Rightarrow CB \Rightarrow CBO$ and no FDs can help so $\{C^+\} \;=\;\{C, B, O\}$

$\{ CS^+ \}: CS \Rightarrow CSQ \Rightarrow CSQR \Rightarrow CSQRB \Rightarrow CSQRBO$ and no other FDs can help so $\{CS^+\} \;=\; \{C, S, Q, R, B, O\}$

These are all the closures that we could get from FDs. We found that one of them leads to a superkey (also a key in this case). That being $\{CS^+\}$.  
All other single attribute closures would be trivial, as there is no FD for them to grow beyond just themselves. Further, all other resulting closures would just be unions of the ones stated already.  
From this we can logic out the resulting keys (and superkeys) for this relation...  

$C \wedge S$ both can not be derived from any other attributes and FDs. So they would both have to be present in a key for this relation.  
Every other attribute can be derived from others. Lets see what further assertions we can make about keys for this relation.
$Q$ can only be derived from the presence of both $C \wedge S$. As $C \wedge S$ have to already be present in a key, adding $Q$ would destroy the minimality requirement of the key.
$R$ can be derived from the presence of $S$. As $S$ already has to be present in a key, adding $R$ would destroy the minimality requirement of the key.
$B \wedge O$ can both be derived from the presence of $C$. As $C$ already has to be present in a key, adding $B \vee O$ (or both) would destroy the minimality requirement of the key.

We have noted that $C \wedge S$ are able to derive all other attributes of the relation, and we have also noted that they have to be explicitly present in any superkey since they cant be derived.  
From this we can see that $CS$ is a subset of all superkeys then, and further since removing either $C \vee S$ would no longer result in a superkey - then $CS$ is a key, and in fact it is the only key possible. (Since $CS$ is a subset of every other superkey.)
