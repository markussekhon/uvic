Given two relations $R$ and $S$, where $R$ contains $M$ tuples, $S$ contains $N$ tuples, and $M > N > 0$, give the *minimum* and *maximum* sizes (in tuples) for the relation resulting from each of the expression shown below. 

**For each expression below, your answer must state any assumptions about the schemas** for $R$ and $S$ that are needed to ensure the expression is meaningful. Providing your answer for an expression without any of your stated assumptions may result in a mark of zero for that answer.

1. $R \cup S$
2. $R \cap S$
3. $R - S$
4. $R \times S$
5. $\sigma_{a=5}(R)$
6. $\pi_a(R)$
7. $R \bowtie S$

#### Answers for A
[Instructor only: A mark --]

*Your answers for Question A begin here*

1. $R \cup S$  
**Assumption:** Schemas for both $R$ and $S$ must match - they must have same exactly the same attributes.  
**Min size:** If all tuples in $S$ also exist in $R$, then the (min) size of $R \cup S$ would be $M$ tuples.  
**Max size:** If $R$ and $S$ have no tuples in common, then the (max) size of $R \cup S$ would be $M + N$ tuples.  

2. $R \cap S$  
**Assumption:** Schemas for both $R$ and $S$ must match - they must have same exactly the same attributes.  
**Min size:**  If $R$ and $S$ have no tuples in common, then the (min) size of $R \cap S$ would be $0$ tuples.  
**Max size:**  If all tuples in $S$ also exist in $R$, then the (max) size of $R \cap S$ would be $N$ tuples.

3. $R - S$  
**Assumption:** Schemas for both $R$ and $S$ must match - they must have same exactly the same attributes.  
**Min size:**  If all tuples in $S$ also exist in $R$, then the (min) size of $R - S$ would be $M-N$ tuples.  
**Max size:**  If $R$ and $S$ have no tuples in common, then the (max) size of $R - S$ would be $M$ tuples.

4. $R \times S$  
**Assumption:** No requirements for schemas.  
**Min size:** As schemas are different for tuples in $R$ and $S$ the size of $R \times S$ will be $M \times N$ tuples.  
**Max size:** As schemas are different for tuples in $R$ and $S$ the size of $R \times S$ will be $M \times N$ tuples.  

5. $\sigma_{a=5}(R)$  
**Assumption:** The schema for $R$ must have an attribute $a$.  
**Min size:**  If all tuples have the value that $a \neq 5$, then none of them meet the condition $a = 5$. So the size of $\sigma_{a=5}(R)$ would be $0$ tuples.  
**Max size:**  If all tuples have the value that $a = 5$, then all of them meet the condition $a = 5$. So the size of $\sigma_{a=5}(R)$ would be $M$ tuples.  

6. $\pi_a(R)$  
**Assumption:** The schema for $R$ must have an attribute $a$, and after this operation the only attribute for the tuples that remain would be $a$.  
**Min size:** If all tuples had the same values for all remaining attributes, which in this case would just be $a$, then the size of $\pi_a(R)$ would be $1$ tuple.  
**Max size:**  If all tuples had different values for all remaining attributes, which in this case would just be $a$, then the size of $\pi_a(R)$ would be $M$ tuples.  

7. $R \bowtie S$  
**Assumption:**  There are no requirements for schemas.  
**Min size:** If there are no common values for the common attributes of $R$ and $S$, then the size of $R \bowtie S$ would be $0$ tuples.  
**Max size:** If there are no common attributes for $R$ and $S$, then $R \bowtie S$ is the same as doing $R \times S$, which has a size of $M \times N$ tuples.
